```python
def make_soup(url):
    '''
    This helper function takes in a url and requests and parses HTML
    returning a soup object.
    '''
    url = 'https://github.com/search?o=desc&q=environmental&s=updated&type=Repositories'
    page = rq.get(url).text
    soup = bs(page)
    return soup
```

```python
make_soup(url)
```
